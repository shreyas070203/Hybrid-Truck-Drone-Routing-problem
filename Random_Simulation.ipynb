```python
import numpy as np
from sklearn.metrics import pairwise_distances
import matplotlib.pyplot as plt
from scipy.spatial import ConvexHull
from k_means_constrained import KMeansConstrained

# Function to visualize clusters
def plot_clusters(depot_location, customer_locations, cluster_assignments):
    # Create a colormap with a unique color for each cluster
    cmap = plt.cm.get_cmap('tab10', len(np.unique(cluster_assignments)))

    plt.scatter(depot_location[0], depot_location[1], color='red', marker='D', label='Depot')

    for cluster_num in np.unique(cluster_assignments):
        cluster_points = customer_locations[cluster_assignments == cluster_num]
        centroid = np.mean(cluster_points, axis=0)  # Calculate the centroid of the cluster
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {cluster_num + 1}', cmap=cmap, alpha=0.7)

        # Draw convex hull around the cluster points
        hull = ConvexHull(cluster_points)
        for simplex in hull.simplices:
            plt.plot(cluster_points[simplex, 0], cluster_points[simplex, 1], color='black', linewidth=1)

        plt.annotate(f'Cluster {cluster_num + 1}', centroid, color='black', weight='bold',
                     fontsize=8, ha='center', va='center')

    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title('K-Means Clustering of Customer Locations')
    #plt.legend()
    plt.show()

# K-Means Clustering Function using KMeansConstrained
def kmeans_constrained_clustering(depot_location, customer_locations, drone_ranges, max_drones):
    # Combine customer locations and depot location
    all_locations = np.concatenate(([depot_location], customer_locations), axis=0)

    # Initialize k-means with 1 cluster
    max_clusters = min(len(customer_locations), max_drones + 1)
    kmeans = KMeansConstrained(n_clusters=11, size_max=max_drones, random_state=0).fit(all_locations)

    while kmeans.n_clusters < max_clusters:
        # Calculate distance between cluster focal point and each assigned customer location
        distances = pairwise_distances(customer_locations, kmeans.cluster_centers_)

        # Check if distance is within the flying range
        within_range = distances <= drone_ranges[:, np.newaxis]

        # Check if the number of locations per cluster is less than the drones carried by the truck
        locations_per_cluster = np.sum(within_range, axis=0)
        valid_clusters = locations_per_cluster <= max_drones

        # If both conditions are satisfied, break the loop
        if np.all(valid_clusters):
            break

        # Increment the number of clusters
        kmeans.n_clusters += 1
        kmeans.fit(all_locations)

    # Enforce size constraint
    while not np.all(np.bincount(kmeans.labels_[1:]) <= max_drones):
        # Identify points belonging to clusters exceeding size_max
        exceeding_clusters = np.where(np.bincount(kmeans.labels_[1:]) > max_drones)[0]

        # Remove points from exceeding clusters
        for cluster in exceeding_clusters:
            cluster_points = np.where(kmeans.labels_ == cluster)[0]
            all_locations = np.delete(all_locations, cluster_points, axis=0)

        # Refit KMeans
        kmeans = KMeansConstrained(n_clusters=1, size_max=max_drones, random_state=0).fit(all_locations)

    # Get cluster assignments for each location
    cluster_assignments = kmeans.predict(all_locations)

    # Return the cluster assignments
    return cluster_assignments[1:]  # Exclude the depot location

# Sample data generation
np.random.seed(42)  # for reproducibility
depot_location = np.array([30, 40])  # Replace with actual depot coordinates

# Generate 100 random customer locations
num_customers = 100
customer_locations = np.random.uniform(low=[25, 35], high=[35, 45], size=(num_customers, 2))

# Generate random drone ranges for each customer
drone_range = np.random.uniform(low=5, high=15)
drone_ranges = np.full(num_customers, drone_range)

# Maximum number of drones carried by the truck
max_drones = 10

# Run the clustering algorithm
cluster_assignments = kmeans_constrained_clustering(depot_location, customer_locations, drone_ranges, max_drones)

# Visualize the clusters
plot_clusters(depot_location, customer_locations, cluster_assignments)
```
